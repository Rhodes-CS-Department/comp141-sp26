# Final Project Write-Up, Including Detailed Specification and Functional Decomposition

When you present your project, you will also turn in a **minimum of one and maximum of three page** write-up. A draft of that write-up is due the last week of class, but the final version will be turned in when you present your project. It builds on the details you have been accumulating all semester and should have the following sections:

#### Overview

This is where you started and you already sent this to me in an email. That can be the starting point, but don't just copy and paste. Some of the particulars may have changed in the interim.

#### Storyboard

You presented this in class. The key is to imagine that you're using your program. What do you want to see? What do you want it to do? How do you want it to work? Write down the user experience that you visualized in your head. I don't want just copied and pasted powerpoint slide pictures. Again, details may have changed. As a reminder, a storyboard contains:
* Step-by-step program flow.
* What the user sees.
* What the user inputs.
* What the program does with that input.
* What the program outputs.

#### Detailed Specification

This is the last stop before you begin coding. This is where the application is fully fleshed-out. The details will depend on your particular project, but you should have figured everything out about what your program is going to do when you start coding. Examples include:
* Characters
* Objects
* Data Sources
* Maps
* Puzzles
* Analyses

#### Functional Decomposition

This is where you break the application up into bite-size pieces. It's similar to the multi-step labs you've worked on this semester. Example functions include:
* Check for Win/Loss
* Load Data
* Run an Analysis
* Run a Different Analysis
* Get User Command
* Make a Move
* Update Status

It is easy when you're thinking about a project to make it very large and expansive. I like seeing the creativity, but this is not intended to be a summers-long internship project where you're working on it for eight hours a day for three months. You should, though, expect to spend two full days (16 hours plus or minus) completeing it. This means that you may have to focus the scope. Maybe just a couple of rooms or characters or analyses. **The main point is for you to develop a practical application that demonstrates an understanding of the concepts covered this semester.**

Be sure to highlight scope changes.

### Final Project Presentation and Write-Up (Reminder)

The Final Project will be presented during Finals week on the afternoon of Thursday, May 7th. Each presentation will be 15 minutes and will incude:
1. Overview of the project and runtime details
2. Live demonstration of the program by you
