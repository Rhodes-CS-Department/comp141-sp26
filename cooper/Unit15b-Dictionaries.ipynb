## A Dictionary Can Contain Dictionaries
* Data storage structures like XML, JSON, and Python dictionaries organize information as nested key–value pairs, allowing complex relationships to be represented in a hierarchical format. They make data both human-readable and machine-processable, enabling systems to store, exchange, and interpret structured information consistently.

```python
classDictionary = {"CS-21" : {"Title":"Intro to Comptuer Science", "Days":"MWF", "Time":"8:00-8:50"},
                   "CS-101": {"Title":"Data Structures", "Days":"TTh", "Time":"10:00-11:40"},
                   "CS-103": {"Title":"Computer Science Theory", "Days":"MWF", "Time":"9:00-9:50"}}

print(classDictionary["CS-21"])
print(classDictionary["CS-21"]["Title"])
print(classDictionary["CS-103"]["Time"])
```
