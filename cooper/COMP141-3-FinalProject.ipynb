### Assessment

The video presentation must be between 5 and 10 minutes in length, and include:
* the details from the proposal (title, description, and narrative)
* a walk-through of the code, describing the data structures, functions, and program flow
* a demonstration of the fucntioning of the program

You will also submit a copy of the program and any supporting files (either via email or okpy, I haven't decided yet) so that I can try it (and try to break it).

An 'A' project will have non-trivial objectives, implement those proposed objectives, be presented neatly, clearly, and completely demonstrating an understanding of the concepts, and function without breaking (too easily) when I try it.
'B' and 'C' projects will have objectives that aren't implemented (or aren't implemented as intended), a sloppy presentation, not so clear understanding of the concepts, and/or breaks easily.

**Do your own work. Do not use AI.**

# Some Project Ideas:

### Text-Based Adventure Game (w/ Graphics)

Design and implement a short, simple text-based adventure game. The player enters prompts into the text field (this would be the `main` game loop) related to what they want to do (`look around`, `fight`, `go to`, etc). There is one or several ways to win or lose. The player can visit several locations and take several actions. Every "level" has a corresponding graphical scene.

### Sports (or other) Data Analysis

Design and implement a "sports statistics calculator", which takes in a single `txt` file (for a single player, single team, single season, etc) and computes various statistics (batting average, best time, worst game, etc). The user decides what file to open and what statistic to calculate.

### Image Editor

Build off of Project 8 to implement a more robust image editor. Changing pixel values is the way all image editors work. Possible functions include blacking out a circle or square of pixels, adding stripes to the image, and blurring the image by shuffling pixels around in a neighborhood.
