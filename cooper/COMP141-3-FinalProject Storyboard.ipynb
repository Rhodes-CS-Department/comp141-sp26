### Storyboard

You already have the idea. Now is the time to think through how your program will actually work before you start coding. Professional programmers don't just start typing...they plan first. A storyboard explains:
* Step-by-step program flow.
* What the user sees.
* What the user inputs.
* What the program does with that input.
* What the program outputs.

If someone else read your storyboard, they should be able to understand exactly how your program works even without seeing your code.

#### Storyboard Example

Let's say you're doing an image editor. Your storyboard might look something like this.

**Program Flow**
1. Display a banner welcoming the user to the Fantastic Image Editor (TM).
2. Ask the user to enter the name of a file to edit. Be sure to confirm that the file exists and that it's a proper image file.
3. Display the image.
4. Display a menu of options.
5. Ask the user to select one of the options.
6. Execute the selected option.
7. Display the updated image.
8. Go back to step 4 and display the menu of options.

**Options**
* Add a shape to the image - another menu with the shape to add (square or circle, outline or filled) then prompts for size and location
* Blur the image - prompt for amount of blur
* Recolor the image - prompt for new color mapping
* Save the image - prompt for a new name

The key is to imagine that you're using your program. What do you want to see? What do you want it to do? How do you want it to work? Write down the user experience that you visualized in your head.
