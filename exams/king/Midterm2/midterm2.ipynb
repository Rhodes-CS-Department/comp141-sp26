```python
# part 2
def is_valid(password):
    # check that there are at least 5 letters and at least 1 number
    letters = 0
    numbers = 0
    for c in password:
        if c.isalpha():
            letters += 1
        elif c.isdigit():
            numbers += 1 
        if letters >= 5 and numbers >= 1:
            return True
    return False


def main():
    print(is_valid("password"))     # False
    print(is_valid("p4ssw0rd"))     # True
    print(is_valid("123456789000")) # False
    print(is_valid("mypass1234!"))  # True
    print(is_valid("1234"))         # False

main()
```
