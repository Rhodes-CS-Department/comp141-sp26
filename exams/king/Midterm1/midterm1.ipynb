### Problem 3: Collatz Steps (Challenge)

The **Collatz sequence** works like this: starting from any positive integer `n`:
- If `n` is **even**, divide it by 2
- If `n` is **odd**, multiply it by 3 and add 1
- Keep going until `n` reaches 1

The function `collatz_steps(n)` below is supposed to count and **return** how many steps it takes to reach 1.

But it doesn't work! Can you fix it?

Example: starting from 6: &nbsp; 6 -> 3 -> 10 -> 5 -> 16 -> 8 -> 4 -> 2 -> 1 &nbsp; (8 steps)

```python
# Part 1: Make sure the program runs without error.
# Part 2: Make sure the program does what it's supposed to.
#         Use the description above! Read it very carefully!
def collatz_steps(n)
    steps = 0
    while n == 1:
        if n % 2 = 0:
            n = n // 2
        else:
            n = n * 3
        steps + 1
    print(steps)
```

```python
# tests
print(collatz_steps(1))  # should be 0
print(collatz_steps(4))  # should be 2
print(collatz_steps(6))  # should be 8
```
