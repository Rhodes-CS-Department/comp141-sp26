## Writing Nested Loops
* Design the outer loop without worrying about what goes inside
* Design what goes inside, ignoring the outer loop.
* Put the pieces together, preserving the nesting.

```python
# Nested Loop Example
# Prime Numbers

i = 2
while i < 100: # for every number
    prime = True
    for j in range(2, i): # check if every other number less than it is a factor of it
        if i % j == 0:
            prime = False
            break
    if prime:
        print(i, "is prime")
    i = i + 1
```

```python
# Nested Loop Example
# Printing out all the cards in a card deck

for i in range(4):
    if i == 0:
        suit = "Diamonds"
    elif i == 1:
        suit = "Clubs"
    elif i == 2:
        suit = "Hearts"
    else:
        suit = "Spades"

    for j in range(1,14):
        if j == 1:
            card = "Ace"
        elif j == 11:
            card = "Jack"
        elif j == 12:
            card = "Queen"
        elif j == 13:
            card = "King"
        else:
            card = j

        print(card, "of", suit)
```

```python
# Nested Loop Example
# Printing out all the cards in a card deck
# Let's modify this to print the cards out in descending point value order (aces high)

for i in range(4):
    if i == 0:
        suit = "Diamonds"
    elif i == 1:
        suit = "Clubs"
    elif i == 2:
        suit = "Hearts"
    else:
        suit = "Spades"

    for j in range(14,1,-1):
        if j == 14:
            card = "Ace"
        elif j == 11:
            card = "Jack"
        elif j == 12:
            card = "Queen"
        elif j == 13:
            card = "King"
        else:
            card = j

        print(card, "of", suit)
```
