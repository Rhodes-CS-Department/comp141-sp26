Here's an example run of the program. 
```
Let's calculate the amount of taxes you owe based on your filing status and taxable income.

```



````python
# Write your program in this cell. Be sure to test your program!
print("Let's calculate the amount of taxes you owe based on your filing status and taxable income.")
print("Please enter these values when prompted. You'll need to enter your status as \"Single\" or \"Married\".")
status = input("Enter your tax filing status")
income = float(input("Enter your taxable income in US dollars."))

if status == "Single":
    if income <= 100525:
        tax_owed = income * 0.1
        print("You owe", tax_owed, "in taxes.")
    else:
        tax_owed = 100525 * 0.1 + (income - 100525) * 0.24
        print("You owe", tax_owed, "in taxes.")
elif status == "Married":
    if income <= 201050:
        tax_owed = income * 0.1
        print("You owe", tax_owed, "in taxes.")
    else:
        tax_owed = 201050 * 0.1 + (income - 201050) * 0.24
        print("You owe", tax_owed, "in taxes.")
else:
    print("Invalid filing status")
````
